```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
plt.style.use('seaborn-v0_8-whitegrid')

# Load ALL datasets from GitHub
BASE_URL = "https://raw.githubusercontent.com/buildLittleWorlds/yeller-quarry-data-science/main/data/"

creatures = pd.read_csv(BASE_URL + "creatures.csv")
crews = pd.read_csv(BASE_URL + "crews.csv")
traps = pd.read_csv(BASE_URL + "traps.csv")
catches = pd.read_csv(BASE_URL + "catches.csv")
prices = pd.read_csv(BASE_URL + "prices.csv")
incidents = pd.read_csv(BASE_URL + "incidents.csv")

# Convert dates
catches['date'] = pd.to_datetime(catches['date'])
prices['date'] = pd.to_datetime(prices['date'])
incidents['date'] = pd.to_datetime(incidents['date'])
traps['deployment_date'] = pd.to_datetime(traps['deployment_date'])
traps['retrieval_date'] = pd.to_datetime(traps['retrieval_date'])

# Calculate derived columns
catches['total_value'] = catches['quantity'] * catches['price_per_unit']

print("All Quarry data loaded.")
print(f"  Creatures: {len(creatures)}")
print(f"  Crews: {len(crews)}")
print(f"  Traps: {len(traps)}")
print(f"  Catches: {len(catches)}")
print(f"  Prices: {len(prices)}")
print(f"  Incidents: {len(incidents)}")
```

## Part 5: Yeller Pattern Analysis

Yeller groups form in 2, 3, 5, 7, and rarely 11. Do any numbers in the investigation follow this pattern?

```python
yeller_numbers = [2, 3, 5, 7, 11]

# Number of children missing
print(f"Children missing: {len(boundary_cases)}")
print(f"Is this a yeller number? {len(boundary_cases) in yeller_numbers}")
```

```python
# Days between disappearances
intervals = boundary_cases_sorted['days_since_last'].dropna().tolist()
print(f"Day intervals: {intervals}")
print(f"Are these yeller numbers? {[i in yeller_numbers for i in intervals]}")
```

```python
# Analyze catch quantities for yeller patterns
catches['is_yeller_qty'] = catches['quantity'].isin(yeller_numbers)
yeller_catches = catches[catches['is_yeller_qty']]
print(f"Catches with yeller quantities: {len(yeller_catches)} of {len(catches)}")
print(f"Percentage: {len(yeller_catches) / len(catches) * 100:.1f}%")
```

## Part 6: Cross-Reference Analysis

Look for connections between disappearances and other events.

```python
# What else happened on or around disappearance dates?
for _, case in boundary_cases.iterrows():
    date = case['date']
    print(f"\n=== {date.strftime('%Y-%m-%d')} ===")
    print(f"Disappearance: {case['description']}")
    
    # Catches within 3 days
    nearby_catches = catches[(catches['date'] >= date - pd.Timedelta(days=3)) & 
                            (catches['date'] <= date + pd.Timedelta(days=3))]
    if len(nearby_catches) > 0:
        print(f"Catches within 3 days: {len(nearby_catches)}")
        
    # Other incidents same day
    same_day = incidents[(incidents['date'] == date) & (incidents['incident_id'] != case['incident_id'])]
    if len(same_day) > 0:
        print(f"Other incidents same day: {len(same_day)}")
        for _, inc in same_day.iterrows():
            print(f"  - {inc['incident_type']}: {inc['description'][:50]}...")
```
